```python

#LOADING THE DATASET
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset('titanic')
print(df.head())
```

UNDERSTOOD THE DATA NOW PLOTTING
1. UNIVARIATE

```python
# AGE -> NUMERICAL -> HISTOGRAM
plt.hist(df['age'])
plt.xlabel("Age")
plt.ylabel("No of Passengers")
plt.title("Age Distribution")
plt.show()

#BOX PLOT TO CHECK EXTREME VALUES AND OUTLIERS
plt.figure(figsize=(6,4))
plt.boxplot(df['age'].dropna())
plt.ylabel("Age")
plt.title("Age Distribution")
plt.show()
```

USING SNS

```python
#using sns
plt.figure(figsize=(6,4))
sns.histplot(df['age'].dropna(),kde=True)
plt.xlabel("Age")
plt.ylabel("No of Passengers")
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(df['age'].dropna())
plt.xlabel("AGE")
plt.show()
```

CHECK RELATIONSHIP OF SEX VS AGE

```python
#boxplot
#1. using matplotlib

plt.figure(figsize=(6,4))
plt.boxplot([df[df['sex']=='male']['age'].dropna(),df[df['sex']=='female']['age'].dropna()])
plt.xlabel("Sex")
plt.ylabel("Age")
plt.xticks([1,2],['Male','Female'])
plt.show()

#using sns
plt.figure(figsize=(6,4))
sns.boxplot(x='sex',y='age',data=df)
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

RELATIONSHIP BETWEEN SURVIVAL / AGES

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='survived', y='age', data=df)
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```

FARE VS SURVIVAL

```python
#FARE IS NUMERICAL -> HISTOGRAM
import numpy as np
df['fare_log'] = np.log1p(df['fare'])
plt.figure(figsize=(8,5))
sns.histplot(df['fare'],bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(x='survived',y='fare',data=df)
plt.ylabel("fare paid")
plt.xlabel("Survived")
plt.show()
```

```python
sns.boxplot(x='pclass', y='fare', data=df)
plt.show()
sns.countplot(x='pclass', hue='survived', data=df)
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.show()
```

```python
sns.catplot(x='pclass', hue='survived', col='sex',
            kind='count', data=df)
plt.show()
```
