LABEL ENCODING

Use when:

Feature is truly ordinal (but better to use OrdinalEncoder)

Tree-based models (Random Forest, Decision Tree) can sometimes handle it

```python
from sklearn.preprocessing import LabelEncoder
data = ['Red', 'Green', 'Blue', 'Red']
le = LabelEncoder()
encoded_data = le.fit_transform(data)
print(f"Encoded Data: {encoded_data}")
```

ONE HOT ENCODING

Use when:

Categories are nominal (no order)

Number of categories is small

Linear models (Logistic Regression, SVM)

```python
import pandas as pd
data = ['Red', 'Blue', 'Green', 'Red']
df = pd.DataFrame(data, columns=['Color'])
one_hot_encoded = pd.get_dummies(df['Color'])
print(one_hot_encoded)
```

ORDINAL ENCODING : You must define order manually.

Use when:

Feature has natural ranking:

Education level

Size (Small, Medium, Large)

Satisfaction rating

```python
from sklearn.preprocessing import OrdinalEncoder
data = [['Low'], ['Medium'], ['High'], ['Medium'], ['Low']]
encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
encoded_data = encoder.fit_transform(data)

print(f"Encoded Ordinal Data: {encoded_data}")
```

BINARY ENCODING : Convert category в†’ number

Convert number в†’ binary

Split binary digits into columns

Use when:

Too many categories

One-hot creates too many columns

```python
import category_encoders as ce
data = ['Red', 'Green', 'Blue', 'Red']
encoder = ce.BinaryEncoder(cols=['Color'])
encoded_data = encoder.fit_transform(pd.DataFrame(data, columns=['Color']))
print(encoded_data)
```

FREQUENCY ENCODING

Use when:

Many categories

Quick lightweight encoding

```python
import pandas as pd
data = ['Red', 'Green', 'Blue', 'Red', 'Red']
series_data = pd.Series(data)
frequency_encoding = series_data.value_counts()

encoded_data = [frequency_encoding[x] for x in data]
print("Encoded Data:", encoded_data)
```

| Encoding  | Creates Order?        | Expands Columns? | Best For                 |
| --------- | --------------------- | ---------------- | ------------------------ |
| Label     | вќЊ (but acts like yes) | No               | Tree models              |
| One-Hot   | вќЊ                     | Yes              | Nominal small categories |
| Ordinal   | вњ…                     | No               | Ranked categories        |
| Target    | Based on target       | No               | High-cardinality         |
| Binary    | No                    | Moderate         | Many categories          |
| Frequency | No                    | No               | Large datasets           |


If feature is:

Nominal (Color, City) в†’ One Hot

Ordinal (Low, Medium, High) в†’ Ordinal

Many categories (100+) в†’ Target or Binary

Tree model only в†’ Label can work

Linear model в†’ One Hot preferred
